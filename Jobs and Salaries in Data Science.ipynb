```python
import matplotlib.pyplot as plt
```

```python
#Average Salaries for each year
average_salaries_by_year = df.groupby('work_year')['salary_in_usd'].mean()
average_salaries_by_year.plot(kind='bar', xlabel='Year', ylabel='Average Salary (USD)', title='Average Salaries by Year')
plt.show()
```

```python
top_job_titles = df['job_title'].value_counts().head(25)
top_job_titles.plot(kind='bar', xlabel='Job Title', ylabel='Frequency', title='Top Job Titles')
plt.show()
```

```python
top_10_jobs = df['job_title'].value_counts().head(10)
top_10_jobs.plot(kind='bar', xlabel='Job Title', ylabel='Frequency', title='Top 10 Most Common Data Jobs')
plt.show()
```

```python
average_salaries_by_title = df.groupby('job_title')['salary_in_usd'].mean()
top_job_titles_highest_salary = average_salaries_by_title.nlargest(5)  # Get top 5 job titles with highest average salaries
top_job_titles_lowest_salary = average_salaries_by_title.nsmallest(5)   # Get top 5 job titles with lowest average salaries
top_job_titles_highest_salary.plot(kind='bar', xlabel='Job Title', ylabel='Average Salary (USD)', title='Top Job Titles with Highest Average Salaries')
plt.show()
top_job_titles_lowest_salary.plot(kind='bar', xlabel='Job Title', ylabel='Average Salary (USD)', title='Top Job Titles with Lowest Average Salaries')
plt.show()
```

```python
common_job_titles = df['job_title'].value_counts().head(5).index
average_salaries_common_titles = df[df['job_title'].isin(common_job_titles)].groupby('job_title')['salary_in_usd'].mean()
average_salaries_common_titles.plot(kind='bar', xlabel='Job Title', ylabel='Average Salary (USD)', title='Average Salaries for the Most Common Job Titles')
plt.show()
```

```python
job_category_distribution = df['job_category'].value_counts().head(10)  # Assuming you want to visualize the top 10 job categories
job_category_distribution.plot(kind='barh', xlabel='Frequency', ylabel='Job Category', title='Distribution of Job Categories')
plt.show()
```

```python
average_salary_by_category = df.groupby('job_category')['salary_in_usd'].mean().sort_values(ascending=False)
average_salary_by_category.plot(kind='bar', xlabel='Job Category', ylabel='Average Salary (USD)', title='Average Salaries by Job Category')
plt.show()
```

```python
average_salary_by_currency = df.groupby('salary_currency')['salary_in_usd'].mean().sort_values(ascending=False)
average_salary_by_currency.plot(kind='bar', xlabel='Currency', ylabel='Average Salary (USD)', title='Average Salary in USD by Currency')
plt.show()
```

```python
average_salary_by_experience = df.groupby('experience_level')['salary_in_usd'].mean().sort_values(ascending=False)
average_salary_by_experience.plot(kind='bar', xlabel='Experience Level', ylabel='Average Salary (USD)', title='Average Salary by Experience Level')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Create the grouped bar plot
plt.figure(figsize=(12, 8))

# Visualize average salary based on employment type, work setting, and company size
ax = sns.barplot(x='employment_type', y='salary_in_usd', hue='work_setting', data=df)

# Set labels and title
plt.xlabel('Employment Type')
plt.ylabel('Average Salary (USD)')
plt.title('Average Salary by Employment Type, Work Setting, and Company Size')

# Show the plot
plt.tight_layout()
plt.show()
```
